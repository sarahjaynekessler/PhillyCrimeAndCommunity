```python
def getridofdash(add):
    add_0 = add.split('-')[0]
    
    for i in [add.split('-')[-1]]:
        words = i.split(' ')[1:]
        add_1 = ' '.join(words)
        
    cleaned_add = add_0+' '+add_1
    return(cleaned_add)
    
```

# start final cleaning and merge of all datasets

```python
crime = pd.read_csv('crime_recent_zip.csv')
```

```python
hours = []
for i in crime.hour_:
    try:
        hours.append(str(int(i)))
    except:
        hours.append(i)
```

```python
crime['hour'] = hours
crime= crime.drop(columns = {'hour_','lat','lng','dc_dist'})
crime.head()
```

```python
ohecols = ['text_general_code','hour']
one_hot = pd.get_dummies(crime[ohecols])
crime = crime.drop(ohecols,axis=1)
crime = crime.join(one_hot)
```

```python
zips = []
for i in crime.zipcode:
    try:
        zips.append(str(int(i)))
    except:
        zips.append(np.nan)
crime.zipcode = zips
crime.head()
```

```python
crime.dispatch_date = pd.to_datetime(crime.dispatch_date)
crime.head()
```

```python
crime['num_crimes']=1
per = crime.dispatch_date.dt.to_period("M")
crime_grouped = crime.groupby(['zipcode',per])[crime.columns[2:]].count()\
    .reset_index()
crime_grouped.to_csv('../datasets/cleaned_data/crime_grouped.csv',index=False)
crime_grouped
```

```python
crime_merge = crime_grouped.merge(art_zips,on='zipcode',how='left').merge(fm_zips,on='zipcode',how='left')\
    .merge(parks_zips,on='zipcode',how='left')\
    .merge(landmark_zips,on = 'zipcode',how='left')\
    .merge(housing_zips,on = 'zipcode',how='left')\
    .merge(nac_office_zips,on = 'zipcode',how='left')\
    .merge(nac_zips,on = 'zipcode',how='left')
```

```python
crime_merge.to_csv('../datasets/final_df/crime_total.csv',index=False)
```
